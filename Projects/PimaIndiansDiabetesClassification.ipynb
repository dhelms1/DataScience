# Problem:
* Can we predict the onset of diabetes within 5 years in Pima Indians given medical details? The baseline performance of predicting the most prevalent class is a classification accuracy of approximately 65%. Top results achieve a classification accuracy of approximately 77%.

# Data:
* 769 observations with 8 input varaibles and 1 ouytput varialbe (missing values are encoded with 0 values). An important note is that all observations are females above the age of 21.

   1. Number of times pregnant (ntp)
   2. Plasma glucose concentration a 2 hours in an oral glucose tolerance test (pgc)
   3. Diastolic blood pressure (mm Hg) (dbp)
   4. Triceps skin fold thickness (mm) (tsft)
   5. 2-Hour serum insulin (mu U/ml) (2hsi)
   6. Body mass index (weight in kg/(height in m)^2) (bmi)
   7. Diabetes pedigree function (dpf)
   8. Age (years)
   9. Class variable (0 or 1)

# Read in data, create column names, and inspect data
* Class = 0 implies 'tested negative for diabetes'
* Class = 1 implies 'tested positive for diabetes'

Note that the samples are not evenly weighted. The negative observations (class 0) is nearly twice the size of the positive observations (class 1).

```python
columns = ['ntp', 'pgc', 'dbp', 'tsft', '2hsi', 'bmi', ' dpf', 'age', 'class']

df = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.csv', 
                  header=None, names=columns)


print(df['class'].value_counts())
df.head()
```

# Modeling (split/normalize data & test multiple models)

```python
np.random.seed(42) # Reproducible Results

X = df.drop('class', axis=1)
y = df['class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Normalize data
X_train, X_test = data_normalize(X_train, X_test)

models = {'LogisticRegression': LogisticRegression(),
          'KNeighborsClassifier': KNeighborsClassifier(),
          'LinearSVC': LinearSVC(),
          'RandomForestClassifier': RandomForestClassifier()}

# Test models
model_scores = fit_and_score(models, X_train, X_test, y_train, y_test)
model_scores
```

### Tune Hyperparamters using GridSearchCV

We will focus on the LogisticRegression model since it is has the highet accuracy score and will be the quickest to run/tune. Hyperparameter tuning was performed for LinearSVC and RandomForestClassifier, but neither resulted in a higher score than the default LogisticRegression model (and were left out of this final analysis report).

```python
# Tune LogisticRegressionModel
np.random.seed(42)
log_grid = {'C': np.linspace(.9, 1.2, num=200),
            'penalty': ['l2'],
            'solver': ['liblinear']}

gs_log_model = GridSearchCV(LogisticRegression(), 
                            param_grid=log_grid, 
                            cv=5, 
                            verbose=True)

gs_log_model.fit(X_train, y_train)
print(gs_log_model.best_params_)
gs_log_model.score(X_test, y_test)
```

### Evaluate the Model

The above change in C value resulted in an increase in accuracy of 0.012987 (approx 1.3%). The value was found by multiple tests always having a C value in the range [1, 1.3], so a larger sample space in this range resulted in a higher accuracy. However, after evaulating this tuned model, all cross-valdiated metrics were lower than the default LogisticRegression model. From this, we see that the increase in accuracy for the tuned model was strictly on the current data split, and therefore was not used for the final model.

```python
np.random.seed(42)

log_reg = LogisticRegression() # final model with default parameters (best cross-validated metrics scores)
log_reg.fit(X_train, y_train)

y_preds = log_reg.predict(X_test)

plot_roc_curve(log_reg, X_test, y_test)
confusion_heatmap(y_test, y_preds)
```
